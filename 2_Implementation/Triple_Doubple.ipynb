```python
import pandas as pd
```

```python
stats_df = pd.read_csv("2_Season_DF/Derrick_Rose_Playing_Time_Stats.csv")
stats_df.info()
```

```python
#finding null values in the Dataframe
pd.isnull(stats_df).sum()
stats_df.columns
```

```python
stats_df[['Date','TRB', 'AST', 'STL', 'BLK','PTS']]
```

```python
def FindTrpDoub(lst):
    count = 0
    for value in lst:
        if value >=10:
            count+=1
    if count >= 3:
        return "Yes"
    else:
        return "No"  
    
trp_doub_apply = stats_df.apply(lambda row:FindTrpDoub(row[['TRB', 'AST', 'STL', 'BLK','PTS']]),axis=1)
stats_df["Trp Dbl"] = trp_doub_apply
```

```python
stats_df.loc[stats_df["Trp Dbl"] == "Yes"]
```

```python
stats_df.to_csv("2_Season_DF/Derrick_Rose_Playing_Time_Stats.csv")
```
